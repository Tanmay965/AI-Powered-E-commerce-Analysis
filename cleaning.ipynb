### handle null values

#### 1) Customers

```python
customers["customer_zip_code"] = customers["customer_zip_code"].fillna(customers["customer_zip_code"].median())
```

```python
customers["customer_city"] = customers["customer_city"].fillna("Unknown")
```

```python
customers["signup_date"] = customers["signup_date"].fillna(customers["signup_date"].mode()[0])
```

#### 2) Orders

```python
orders["order_delivered_carrier_date"] = orders["order_delivered_carrier_date"].fillna(orders["order_estimated_delivery_date"])
```

```python
orders["order_delivered_customer_date"] = orders["order_delivered_customer_date"].fillna(orders["order_estimated_delivery_date"])
```

#### 3) Order items

```python
order_items["price"] = order_items["price"].fillna(order_items["price"].median())
```

#### 4) Payments table

```python
payments["payment_type"] = payments["payment_type"].fillna("unknown")
```

#### 5) Products table

```python
products["product_category_name"] = products["product_category_name"].fillna("unknown")
```

```python
products["product_length_cm"] = products["product_length_cm"].fillna(products["product_length_cm"].median())
```

```python
products["product_height_cm"] = products["product_height_cm"].fillna(products["product_height_cm"].median())
```

```python
products["product_width_cm"] = products["product_width_cm"].fillna(products["product_width_cm"].median())
```

#### 6) Reviews table

```python
reviews["review_comment_title"] = reviews["review_comment_title"].fillna("No Title")
```

```python
reviews["review_comment_message"] = reviews["review_comment_message"].fillna("No Comment")
```

```python
reviews["review_answer_timestamp"] = reviews["review_answer_timestamp"].fillna(reviews["review_creation_date"])
```

#### 7) Sellers

```python
sellers["seller_zip_code"] = sellers["seller_zip_code"].fillna(sellers["seller_zip_code"].median())
```

```python
sellers["seller_city"] = sellers["seller_city"].fillna("Unknown")
```

#### 8) Geolocation table

```python
geolocation["geolocation_lat"] = geolocation["geolocation_lat"].fillna(geolocation["geolocation_lat"].median())
```

```python
geolocation["geolocation_lng"] = geolocation["geolocation_lng"].fillna(geolocation["geolocation_lng"].median())
```

```python
geolocation["geolocation_city"] = geolocation["geolocation_city"].fillna("Unknown")
```

## outlier handling

#### customers

```python
customers = customers[
    (customers["age"] >= 18) &
    (customers["age"] <= 80)
]                                 # invalid ages
```

#### payments

```python
payments = payments[
    payments["payment_installments"] >= 0
]                                             # negative installments
```

```python
payments = payments[
    payments["payment_value"] >= 0
]                                            # negative payments values
```

#### products

```python
products = products[
    (products["product_weight_g"] > 0) &
    (products["product_weight_g"] <= 50000)
]                                                 # Invalid Product Weights
```

```python
products = products[
    products["product_height_cm"] > 0
]                                              # negative height
```

#### sellers

```python
sellers = sellers[
    sellers["seller_zip_code"] > 0
]                                          # negative ZIP codes
```

#### geolocation

```python
geolocation = geolocation[
    (geolocation["geolocation_lat"] >= -90) &
    (geolocation["geolocation_lat"] <= 90)
]                                                 # latitude outside valid Earth range
```

```python
geolocation = geolocation[
    (geolocation["geolocation_lng"] >= -180) &
    (geolocation["geolocation_lng"] <= 180)
]                                                 # longitude outside valid Earth range
```

#### reviews

```python
reviews = reviews[
    (reviews["review_score"] >= 1) &
    (reviews["review_score"] <= 5)
]                                         # scores above 5
```
